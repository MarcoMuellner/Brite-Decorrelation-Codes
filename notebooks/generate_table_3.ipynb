```python
from astroquery.exceptions import TableParseError
import warnings

satellite_order = ['BAb','UBr','BTr','BLb','BHr']

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for field in fields:
        print(field)
        stars = load(field)
        result_data = []
        lines = ""
        for star in stars:
            data = star.get_all_data_sets(star.results[0])
            merged_data = [i for i in data if i.combined and "merged" in i.filename]
            tmp_order = []
            for i in satellite_order:
                if i in [j.satellite for j in merged_data]:
                    tmp_order.append([j for j in merged_data if j.satellite == i][0])
            merged_data = tmp_order

            if len(merged_data) == 0:
                continue
            try:
                simbad_name = name_frame[name_frame.iloc[:,3]==f"{star._number}"]['Star'].values[0]
            except IndexError:
                print(star._number)
                print("ERROR")
                simbad_name = "-"
            try:
                line = f"{star._number}\t\t&{simbad_name}\t\t&{merged_data[0].satellite}\t\t&-\t\t&{merged_data[0].datapoints}\t\t&{'%.4f'%(merged_data[0].noise().value*1000)}\t\t&{'%.4f' % (merged_data[0].ptp_scatter*1000)} \\\\" + "\n"
            except IndexError:
                line = f"{star._number}\t\t&{simbad_name}\t\t&{merged_data[0].satellite}\t\t&-\t\t&{merged_data[0].datapoints}\t\t&-\t\t&{'%.4f'%merged_data[0].ptp_scatter} \\\\" + "\n"

            for i in merged_data[1:]:
                line += f"\t\t\t&\t\t\t\t&{i.satellite}\t\t&-\t\t&{i.datapoints}\t\t&{'%.4f'%(i.noise().value*1000)}\t\t&{'%.4f'%i.ptp_scatter}  \\\\" + "\n"

            lines += line
            start_table = "\\begin{table}[htb]\n\caption{Stars observed in Field "+str(field)+".}\n"
            start_table += "\label{tab:field"+str(field)+"}\n\\begin{center}\n%\\begin{scriptsize}\n\\tabcolsep=0.10cm\n\\begin{tabular}{lllrrcc}"
            start_table += "\hline \hline\n"
            start_table += "HD & Other & Sat & $N_{\\rm Adam}$ & $N_{decor}$ & Noise & Scatter \\\\\n"
            start_table += "    & ID    &     &                &             & \multicolumn{2}{c}{(mmag)} \\\\\n"
            start_table += "\hline\n"

            end_table = "\hline\n\end{tabular}\n%\end{scriptsize}\n\end{center}\n\end{table}"

            with open("field"+str(field)+".tex",'w') as f:
                table = start_table + lines + "\n" +  end_table
                f.write(table)
```

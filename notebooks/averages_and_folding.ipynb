# Averages and Folding

This notebook shows a basic rundown on how to work with averages and rundowns

```python
#this should always be your starting point on any notebook. Loads the data types
from common import * 
```

```python
# Get yourself some data
stars = load(10)
star = stars[0]
data = star.get_data(star.results[0])
```

```python
#Scattering the lightcurve will overplot the averages
data.scatter()
```

```python
#You can also access just the averages
data.lk_ave.scatter()
```

```python
#Or just the normal one
data.lk.scatter()
```

```python
#with the lk objects you can work as usual
data.lk_ave.to_periodogram().plot()
```

# Folding

```python
# Estimate folding at period of max power
data.to_periodogram().period_at_max_power
```

```python
#This will fold the light curve at a given period, and then plot it
data.fold(period=data.to_periodogram().period_at_max_power).plot()
```
